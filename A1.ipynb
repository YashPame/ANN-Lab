```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def binaryStep(x):
    return np.heaviside(x,1)

plt.figure(figsize=(4, 2))
x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.title('binaryStep Activation Function')
plt.show()
```

```python
def linearFun(x):
    return x

plt.figure(figsize=(4, 2))
x = np.linspace(-10,10)
plt.plot(x, linearFun(x))
plt.title("Linear Activation Function")
plt.show()
```

```python
def sigmoidFun(x):
    return 1 / (1+np.exp(-x))

plt.figure(figsize=(4, 2))
x = np.linspace(-10,10)
plt.plot(x, sigmoidFun(x))
plt.title("Sigmoid Activation Function")
plt.show()
```

```python
def tanhFun(x):
    return np.tanh(x)

plt.figure(figsize=(4, 2))
x = np.linspace(-10,10)
plt.plot(x, tanhFun(x))
plt.title("Tanh Activation Function")
plt.show()
```

```python
def ReluFun(x):
    return np.maximum(0,x)

plt.figure(figsize=(4, 2))
x = np.linspace(-10,10)
plt.plot(x, ReluFun(x))
plt.title("Relu Activation Function")
plt.show()
```
